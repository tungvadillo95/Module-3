```sql
SELECT ProductID,
  COUNT(OrderID) 'Orders Count',
  SUM((1 - Discount) * UnitPrice * Quantity) 'Sales'
FROM [Order Details]
GROUP BY ProductID
ORDER BY Sales DESC
GO

SELECT [O D].ProductID, P.ProductName,
  COUNT([O D].ProductID) 'Orders Count',
  SUM((1 - [O D].Discount) * [O D].UnitPrice * [O D].Quantity) 'Sales'
FROM [Order Details] [O D]
  LEFT JOIN Products P ON [O D].ProductID = P.ProductID
GROUP BY [O D].ProductID, P.ProductName
ORDER BY Sales DESC
GO

SELECT [O D].ProductID, P.ProductName, S.CompanyName Suplier,
  COUNT([O D].ProductID) 'Orders Count',
  SUM((1 - [O D].Discount) * [O D].UnitPrice * [O D].Quantity) 'Sales'
FROM [Order Details] [O D]
  LEFT JOIN Products P ON [O D].ProductID = P.ProductID
  LEFT JOIN Suppliers S ON P.SupplierID = S.SupplierID
GROUP BY [O D].ProductID, P.ProductName, S.CompanyName
ORDER BY Sales DESC
GO

SELECT Suplier, SUM(Sales)
FROM (
  SELECT [O D].ProductID,
    P.ProductName,
    S.CompanyName Suplier,
    COUNT([O D].ProductID) 'Orders Count',
    SUM((1 - [O D].Discount) * [O D].UnitPrice * [O D].Quantity) 'Sales'
  FROM [Order Details] [O D]
         LEFT JOIN Products P ON [O D].ProductID = P.ProductID
         LEFT JOIN Suppliers S ON P.SupplierID = S.SupplierID
  GROUP BY [O D].ProductID, P.ProductName, S.CompanyName) AS Sales
GROUP BY Suplier
ORDER BY SUM(Sales) DESC 
GO

```
